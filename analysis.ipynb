```python
corr = df['Profit_Margin_USD'].corr(df['Resale_Price_USD'])
print(corr)
```

```python
corr = df[['Retail_Price_USD','Resale_Price_USD','Profit_Margin_USD','Days_in_Inventory']].corr()
corr
```

```python
sns.heatmap(corr,annot=True,cmap='coolwarm')
plt.show()
```

```python
df['Sale_Date'] = pd.to_datetime(df['Sale_Date'])
df['Year'] = df['Sale_Date'].dt.year
df['Month'] = df['Sale_Date'].dt.month
```

```python
df.groupby('Year')['Profit_Margin_USD'].mean()
```

```python
sns.lineplot(x='Year',y='Profit_Margin_USD',data = df)
plt.show()
```

```python
df.groupby(['Sales_Channel','Shoe_Model'])['Profit_Margin_USD'].mean()
```

```python
pd.pivot_table(df,values='Profit_Margin_USD',index='Sales_Channel',columns='Shoe_Model',aggfunc='mean')
```

```python

top_models = df.groupby('Shoe_Model')['Profit_Margin_USD'].mean().sort_values(ascending=False).head(5)
top_models
```

```python
fasest_channel = df.groupby('Sales_Channel')['Days_in_Inventory'].mean().sort_values()
fasest_channel
```

```python
sns.boxplot(x='Sales_Channel',y='Profit_Margin_USD',data=df)
plt.show()
```

```python
df['Sale_Date'] = pd.to_datetime(df['Sale_Date'])
df['Month'] = df['Sale_Date'].dt.month
monthly_profit = df.groupby('Month')['Profit_Margin_USD'].mean()
monthly_profit
```

```python
sns.lineplot(x=monthly_profit.index, y=monthly_profit.values)
plt.xlabel('Month')
plt.ylabel('Average Profit')
plt.show()
```

**Conclusion**

In this project, exploratory data analysis was performed on sneaker resale data to derive meaningful business insights related to pricing, profitability, and inventory efficiency.
